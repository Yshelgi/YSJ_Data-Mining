**数据分析和挖掘任务**中经常会遇到高维数据集的情况；比如用于对手写数字进行分类建模的[MNIST图像数据集](https://zhuanlan.zhihu.com/p/36592188 "MNIST图像数据集")。我们通常会把该数据集中图像的每个像素点作为特征进行建模；在这种情况下，MNIST图像数据就是一种典型的高维数据。

那么，探索和理解数据的维度有什么用处呢？我们可以这样考虑：如果两个数据在维度空间分布中的`距离`非常近，甚至可以认为是紧挨在一起，那么我们是不是可以认为这个两个数据所代表的样本（比如两张手写数字的图像）`非常可能`属于同一个类别呢？事实上，数据样本之间在维度空间的`距离`就是我们在机器学习中定义的样本相似性；而这种相似性就是作为样本分类的最重要的依据！




而高维空间的性质和我们通常感知到的三维空间有着很大的区别，有时候甚至是反直觉的(counterintuitive)！比如高维空间里面的数据点之间的`距离`计算就会得到一些很有趣的结果。下面我们就利用Python来探索一下高维空间里面数据点的距离关系。

<hr>

第一个问题：假设在二维平面上有个 $1*1$ 的正方形，我们在这个正方形内部**随机**任取两个点，请问这两个点之间的平均（期望）距离是多少？

我们当然可以用概率论里面的数学公式加上一点点微积分知识来解决这个问题，但是我们同样可以用Python来模拟**（暴力破解）**这个问题得到非常接近的答案（数值解）；比如随机取$n$对数据点，然后计算平均距离。

所以，下面是你要完成的：
- 分别计算 $n = 1,10,100,1000,10000$ 的时候，正方形内部任意两点之间的平均距离 
- 将这些平均距离可视化

你认为第一个问题的答案是多少呢？

```python
import numpy as np
from random import random 

# Your codes here
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

def get_point(dim,n):
    point=np.random.rand((2*n),dim)
    point=np.array(point,dtype='float32')
    return point

def get_dist(dim,n):
    point=get_point(dim,n)
    for i in range(len(point)):
        yield np.sum(np.sum(np.sqrt(np.sum(np.power((point[i]- point[i+1:]),2), axis=1))))/(2*n)

        
def plot_dis():
    dim=int(input("请输入你想要得到的矩阵维度"))
    dist_list=[]
    n=[1,10,100,1000,10000]
    for i in n:
        dist=np.sum([x for x in get_dist(dim,i)])/i
        dist_list.append(dist)
    print(f"n={n}时,平均距离分别为为:{dist_list}")
    plt.figure(figsize=(10,6))
    plt.bar(np.arange(5),dist_list)
    for x,y in zip(np.arange(5),dist_list):
        plt.text(x, y+0.005, '%.2f' % y, ha='center', va= 'bottom',fontsize=11)
    plt.xlabel("n(n为n对数据点)")
    plt.xticks(np.arange(5),['1','10','100','1000','10000'])
    plt.ylabel("任意点间的平均距离")
    plt.title("1*1正方形内任意两点间的距离")
    plt.show()

plot_dis()
```

那么，继续探索下去，三维空间的边长为$1*1*1$立方体中任取两点，其平均距离是多少呢？

```python
# Your codes here
def plot_dis2():
    dim=int(input("请输入你想要得到的矩阵维度"))
    dist_list=[]
    n=[1,10,100,1000,10000]
    for i in n:
        dist=np.sum([x for x in get_dist(dim,i)])/i
        dist_list.append(dist)
    print(f"n={n}时,平均距离分别为为:{dist_list}")
    plt.figure(figsize=(10,6))
    plt.bar(np.arange(5),dist_list)
    for x,y in zip(np.arange(5),dist_list):
        plt.text(x, y+0.005, '%.2f' % y, ha='center', va= 'bottom',fontsize=11)
    plt.xlabel("n(n为n对数据点)")
    plt.xticks(np.arange(5),['1','10','100','1000','10000'])
    plt.ylabel("任意点间的平均距离")
    plt.title("1*1*1正方体内任意两点间的距离")
    plt.show()
plot_dis2()
```

好了，大家可能已经看出一点端倪了... 那么在1,000,000万维的空间中，边长为1的超正方体其中任意两点的平均距离是多少呢？

`
需要格外注意的是，计算以上问题可能需要一点技巧；要不等待很长时间，要么暂用大量内存。
`

```python
# Your codes here
#本来想着还是利用用上面的代码计算，但是好像内存过大无法计算了，还是重新想想怎么弄吧
#重点：减少循环用numpy向量化计算，减少列表尽量使用迭代器
dim=1000000
n=[1,10,100,1000,10000]
res=[]
def calc(n):
    for i in range(n):
        x=np.random.rand(1,dim)
        y=np.random.rand(1,dim)
        yield np.linalg.norm(x-y)
        
for i in n:
    res.append(np.sum([x for x in calc(i)])/i)
print(f"n={n}时,平均距离分别为为:{res}")
plt.figure(figsize=(10,6))
plt.bar(np.arange(5),res)
for x,y in zip(np.arange(5),res):
    plt.text(x, y+0.005, '%.2f' % y, ha='center', va= 'bottom',fontsize=11)
plt.xlabel("n(n为n对数据点)")
plt.xticks(np.arange(5),['1','10','100','1000','10000'])
plt.ylabel("任意点间的平均距离")
plt.title("超正方体内任意两点间的距离")
plt.show()
```
