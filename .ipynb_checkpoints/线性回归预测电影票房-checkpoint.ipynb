***我们发现这里面有缺失值存在，所以接下来我们要处理缺失值***

```python
columns = ['评分','评分人数','想看人数','物料播放量','总票房','首周票房','首日票房']
for i in columns:
    movie_data[i] = movie_data[i].fillna(movie_data[i].mean())
```

## 模型训练

### 1.线性回归模型

```python
# 切割数据集,训练集和测试集按照8:2来划分
X_train,X_test,y_train,y_test = train_test_split(X_std ,y ,test_size = 0.2 ,random_state = 0)
```

```python
# 创建模型,首先使用默认参数
lrg = LinearRegression()
lrg.fit(X_train,y_train)
prediction_test_lrg = lrg.predict(X_test)
```

```python
# 使用R2-score和均方误差来评估模型
model_name = 'LinearRegression'
print(model_name, '| R2-score: %.4f' % r2_score(y_test, prediction_test_lrg))
print(model_name,'| Mean square error: %.4f' % mean_squared_error(y_test, prediction_test_lrg))
```

-   $R^2$<=1
-   $R^2$越大越好。当我们的预测模型不犯任何错误时，$R^2$得到最大值1。
-   当我们的模型等于基准模型时，$R^2$=0
-   如果$R^2$<0说明我们学习到的模型还不如基准模型。此时很有可能我们的数据不存在任何线性关系
- 我们这里的$R^2$ = 0.94，故此模型还是有意义的.
- 均方误差是平均误差的方法，可以评价数据的变化程度，均方误差的值越小，说明该模型描述实验数据的准确度越高。

```python
# 我们查看一下各个变量的系数
coefficients = pd.DataFrame(lrg.coef_,index = X.columns)
coefficients
```
