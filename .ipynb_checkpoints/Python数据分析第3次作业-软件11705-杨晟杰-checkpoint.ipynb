#### 练习题：

（1) 在x的取值范围[0, 2]之间绘制函数  $f(x)=\sin^2(x-2)e^{-x^2}$ <br>
并加上合适的X/Y轴的labels和title.

（2) 按照下面的案例图片来选择合适的subplot组合，并实现所用线条的不同颜色、不同样式、不同marker以及legend和箭头文字的组合



```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']='fangsong'
plt.rcParams['axes.unicode_minus']=False
```

```python
def f(x):
    return (np.sin(x-2)**2)*np.exp(-x**2)

X=np.linspace(0,2,30)
Y=f(X)

plt.figure(figsize=(20,6))
plt.plot(X,Y,color='b')
plt.title("函数示意图$y=f(x)=sin^2(x-2)e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,2)
plt.ylim(0,1)
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(20,16))

ax1 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax1.plot(X,Y,'r*')
plt.title("函数示意图$y=f(x)=sin^2(x-2)e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,2)
plt.ylim(0,1)
plt.grid()


ax2 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax2.plot(X,Y,'m.-.')
plt.title("函数示意图$y=f(x)=sin^2(x-2)e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,2)
plt.ylim(0,1)
plt.grid()


ax3 = plt.subplot2grid((3, 3), (2, 0))
ax3.plot(X,Y,color='g',label="注意legend出现")
plt.title("函数示意图$y=f(x)=sin^2(x-2)e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,2)
plt.ylim(0,1)
plt.legend()
plt.grid()


ax4 = plt.subplot2grid((3, 3), (2, 1))
ax4.plot(X,Y,color='b')
plt.title("函数示意图$y=f(x)=sin^2(x-2)e^{-x^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.annotate('注意箭头', xy=(0.8,0.5), xytext=(1,0.7), arrowprops=dict(arrowstyle='->'))
plt.xlim(0,2)
plt.ylim(0,1)
plt.grid()

plt.show()
```
